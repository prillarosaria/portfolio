```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#membaca file csv
df = pd.read_csv("google_scholar_teknologi.csv")
df.head(10)
```

```python
ai = df[df['article_title'].str.contains("svm|naive bayes|gradient descent|cluster|speech recognition|deep learning|neural network|machine learning|smart|expert system|intelligent games|digital image processing|natural language processing|robot|feature|decision tree|random forest|regression")]
```

```python
len(ai)
```

```python
mulmed = df[df['article_title'].str.contains("multimedia|image|picture|augmented|game|audio|video|animation|graphic|visual")]
```

```python
len(mulmed)
```

```python
SI = df[df['article_title'].str.contains("information system|knowledge management|e-government|project management|digital marketing|e-commerce|e-learning|e-business|business intelligence|enterprise architecture|technopreneurship|information technology infrastructure")]
```

```python
len(SI)
```

```python
bd = df[df['article_title'].str.contains("big data|time series|warehouse|data|visualization|data analysis|data science|preprocessing|data mining|data management|mining|decision support|financial technology|data management|database|computational statistics")]
bd
```

```python
len(bd)
```

```python
jarkom = df[df['article_title'].str.contains("network|hack|cloud|computer forensics|wireless|telecommunication|security|cyber security")] 
```

```python
len(jarkom)
```

```python
rpl = df[df['article_title'].str.contains("web|software|development|mobile|android mobile|android-based|android based|information engineering")]
```

```python
df2 = pd.DataFrame({
   'RPL': [len(rpl.loc[rpl['year'] == 2016]), len(rpl.loc[rpl['year']== 2017]), len(rpl.loc[rpl['year']== 2018]), len(rpl.loc[rpl['year']== 2019]), len(rpl.loc[rpl['year']== 2020])],
   'AI': [len(ai.loc[ai['year'] == 2016]), len(ai.loc[ai['year']== 2017]), len(ai.loc[ai['year']== 2018]), len(ai.loc[ai['year']== 2019]), len(ai.loc[ai['year']== 2020])],
    'Multimedia': [len(mulmed.loc[mulmed['year'] == 2016]), len(mulmed.loc[mulmed['year']== 2017]), len(mulmed.loc[mulmed['year']== 2018]), len(mulmed.loc[mulmed['year']== 2019]), len(mulmed.loc[mulmed['year']== 2020])],
    'Jaringan Komputer': [len(jarkom.loc[jarkom['year'] == 2016]), len(jarkom.loc[jarkom['year']== 2017]), len(jarkom.loc[jarkom['year']== 2018]), len(jarkom.loc[jarkom['year']== 2019]), len(jarkom.loc[jarkom['year']== 2020])],
    'Big Data': [len(bd.loc[bd['year'] == 2016]), len(bd.loc[bd['year']== 2017]), len(bd.loc[bd['year']== 2018]), len(bd.loc[bd['year']== 2019]), len(bd.loc[bd['year']== 2020])],
    'Sistem Informasi': [len(SI.loc[SI['year'] == 2016]), len(SI.loc[SI['year']== 2017]), len(SI.loc[SI['year']== 2018]), len(SI.loc[SI['year']== 2019]), len(SI.loc[SI['year']== 2020])]
    }, index=['2016', '2017', '2018', '2019', '2020'])
df2
```

# Line Chart

```python
lines = df2.plot(marker="o")
sns.set(font_scale=1.0)
lines.legend(loc='best', frameon=False, bbox_to_anchor = (1.00, 0.6))
plt.xlabel("Tahun Publikasi", labelpad=10)
plt.ylabel("Jumlah Publikasi", labelpad=10)
plt.title("Grafik Perkembangan Bidang Tren Topik Penelitian Artikel Jurnal dalam Bidang Teknologi per Tahun", y=1.04, fontsize=14);
```

```python
plt.savefig('Tren_Topik_Per_Tahun.png')
```

# Pie Chart

```python
y = np.array([df2['RPL'].sum(), df2['AI'].sum(), df2['Multimedia'].sum(), df2['Jaringan Komputer'].sum(), 
              df2['Big Data'].sum(), df2['Sistem Informasi'].sum() ])
color = ['#FF75A0', '#FCE38A', '#EAFFD0', '#95E1D3', '#FBC687', '#EB6383']
plt.pie(y, labels = df2.columns.values , autopct='%1.1f%%', colors = color)
plt.title('Perbandingan Jumlah Artikel per KBK', fontsize=15)
plt.show()
```

# Heatmap

```python
sns.heatmap(df2, cmap='RdPu' , linecolor="black", lw=1, annot=True)
plt.title('Heatmap Jumlah Artikel per KBK dan Tahun', fontsize=15, y = 1.05)
plt.xlabel('KBK', fontsize=14)
plt.ylabel('Tahun', fontsize=14)
```
